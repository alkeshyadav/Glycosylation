```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import pandas as pd 
```

# Figure 1a

```python
# Planaria data - Mass Spec, 3 and 20 GMM plots

df = pd.read_excel('../../Data/Glycan_profile/ms_data/planaria.xlsx',sheet_name='3. 35 & 50 A')
mass_P = df['Centroid Mass'].to_numpy()
Intensity_P = df['Relative Intensity'].to_numpy()
Intensity_P = Intensity_P/Intensity_P.max()

gmm3_P = np.genfromtxt('../../Data/Glycan_profile/GMM_fit_data/planaria_gmm3.txt',delimiter = '\t')
gmm3_P = gmm3_P/gmm3_P.max()

gmm20_P = np.genfromtxt('../../Data/Glycan_profile/GMM_fit_data/planaria_gmm20.txt',delimiter = '\t')
gmm20_P = gmm20_P/gmm20_P.max()
```

```python
# Hydra data - Mass Spec, 3 and 20 GMM plots

df = pd.read_csv('../../Data/Glycan_profile/ms_data/Hydra.txt',delimiter = '\t')
mass_H = df['Centroid Mass'].to_numpy()
Intensity_H = df['Relative Intensity'].to_numpy()
Intensity_H = Intensity_H/Intensity_H.max()

gmm3_H = np.genfromtxt('../../Data/Glycan_profile/GMM_fit_data/hydra_gmm3.txt',delimiter = '\t')
gmm3_H = gmm3_H/gmm3_H.max()

gmm20_H = np.genfromtxt('../../Data/Glycan_profile/GMM_fit_data/hydra_gmm20.txt',delimiter = '\t')
gmm20_H = gmm20_H/gmm20_H.max()
```

```python
# Human Neutrophils - Mass Spec, 3 and 20 GMM plots

x = np.genfromtxt('../../Data/Glycan_profile/ms_data/human_neutrophils1.txt',delimiter="\t")
Intensity_HN = x[1:,1]/x[1:,1].max()
mass_HN = x[1:,0]  

gmm3_HN = np.genfromtxt('../../Data/Glycan_profile/GMM_fit_data/human_neutrophils_3gmm.txt')
gmm3_HN = gmm3_HN/gmm3_HN.max()

gmm20_HN = np.genfromtxt('../../Data/Glycan_profile/GMM_fit_data/human_neutrophils_20gmm.txt')
gmm20_HN = gmm20_HN/gmm20_HN.max()
```

```python
color1 = (107/256,76/256,154/256)
color2 = (62/256,150/256,81/256)
color3 = (204/256,37/256,41/256)
color4 = (146/256,36/256,40/256)
color5 = (54/256,106/256,177/256)
color6 = (218/256,124/256,48/256)
color7 = (148/256,139/256,61/256)

sampled_mass = np.linspace(500,5000,180)
```

```python
mpl.rcParams['axes.linewidth'] = 4 

fig = plt.figure(figsize=(24,12))

color_ms = color6
color_gmm = color5
lw_ms = '2'
lw_gmm = '2' 
ms  = '5'

ax1 = plt.subplot2grid((24,12), (0, 0), rowspan=8, colspan=4)
line1 = ax1.plot(mass_P,Intensity_P, label = 'MSMS Data',linewidth = lw_ms, color = color_ms)
line2 = ax1.plot(sampled_mass,gmm3_P,'-o',label = '3-GMM Fit',linewidth = lw_gmm, markersize = ms, color = color_gmm)
ax1.set_yticks([])
ax1.text(-1400,0.5,'Planaria' ,fontsize = 22,fontweight='bold')
ax1.set_title('3-GMM',fontsize = 22,fontweight='bold')
ax1.legend(fontsize = 16)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.label_outer()

ax2 = plt.subplot2grid((24,12), (8, 0), rowspan= 8, colspan=4)
line1 = ax2.plot(mass_H,Intensity_H, label = 'MSMS Data',linewidth = lw_ms,color = color_ms)
line2 = ax2.plot(sampled_mass,gmm3_H,'-o',label = '3-GMM Fit',linewidth = lw_gmm,markersize = ms, color = color_gmm)
ax2.set_yticks([])
ax2.text(-1400,0.5,'Hydra' ,fontsize = 22,fontweight='bold')
ax2.legend(fontsize = 16)
ax2.label_outer()
ax2.text(-22,0.50,'Relative abundance $\longrightarrow$',rotation='vertical',fontsize = 22,ha='center', va='center')
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)

ax3 = plt.subplot2grid((24,12), (16, 0), rowspan=8, colspan=4)
line1 = ax3.plot(mass_HN,Intensity_HN, label = 'MSMS Data',linewidth = lw_ms,color = color_ms)
line2 = ax3.plot(sampled_mass,gmm3_HN,'-o',label = '3-GMM Fit',linewidth = lw_gmm, markersize = ms, color = color_gmm)
ax3.set_yticks([])
ax3.tick_params(axis='x', which='major', labelsize=16)
ax3.text(-1600,0.5,'  Human \nNeutrophils' ,fontsize = 22,fontweight='bold')
ax3.legend(fontsize = 16)
ax3.text(6000,-0.3,'Glycan Mass/Index $\longrightarrow$',fontsize = 22,ha='center', va='center')
#ax3.arrow(250,-0.3,50,0)
#arrow = mpatches.Arrow(200,-0.3,50,0)
#ax3.add_patch(arrow)
#ax3.text(200,-0.48,'(a)',fontsize = 20,ha='center', va='center')
ax3.spines['right'].set_visible(False)
ax3.spines['top'].set_visible(False)


ax4 = plt.subplot2grid((24,12), (0, 4), rowspan=8, colspan=4)
line1 = ax4.plot(mass_P,Intensity_P, label = 'MSMS Data',linewidth =lw_ms, color = color_ms)
line2 = ax4.plot(sampled_mass,gmm20_P,'-o',label = '20-GMM Fit',linewidth = lw_gmm, markersize = ms,color = color_gmm)
ax4.set_yticks([])
ax4.set_title('20-GMM',fontsize = 22,fontweight='bold')
ax4.legend(fontsize = 16)
ax4.label_outer()
ax4.spines['right'].set_visible(False)
ax4.spines['top'].set_visible(False)

ax5 = plt.subplot2grid((24,12), (8, 4), rowspan=8, colspan=4)
line1 = ax5.plot(mass_H,Intensity_H, label = 'MSMS Data',linewidth =lw_ms,color = color_ms)
line2 = ax5.plot(sampled_mass,gmm20_H,'-o',label = '20-GMM Fit',linewidth = lw_gmm,markersize = ms, color = color_gmm)
ax5.set_yticks([])
ax5.legend(fontsize = 16)
ax5.label_outer()
ax5.spines['right'].set_visible(False)
ax5.spines['top'].set_visible(False)

ax6 = plt.subplot2grid((24,12), (16, 4), rowspan=8, colspan=4)
line1 = ax6.plot(mass_HN,Intensity_HN, label = 'MSMS Data',linewidth = lw_ms, color = color_ms)
line2 = ax6.plot(sampled_mass,gmm20_HN,'-o',label = '20-GMM Fit',linewidth = lw_gmm, markersize = ms, color = color_gmm)
ax6.set_yticks([])
ax6.legend(fontsize = 16)
ax6.tick_params(axis='x', which='major', labelsize=16)
ax6.spines['right'].set_visible(False)
ax6.spines['top'].set_visible(False)


plt.savefig('ms_complexity_gmm.png',format='png',dpi = 400,bbox_inches='tight' )

#fig.tight_layout()
```
